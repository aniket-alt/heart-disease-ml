```python
# === Heart Disease EDA (CRISP-DM Phase 2) — Colab-ready ===
# - Upload or point to a CSV (UCI Heart Disease variants supported)
# - Cleans weird encodings, binarizes target if needed
# - Recode physiologically invalid values -> NaN
# - Outputs: Data dictionary, prevalence & slices, range checks
# - Plots: key histograms & categorical prevalence bars
# - Feature ranking (Mutual Information), numeric correlations
# - Quick baseline logistic regression (holdout) + calibration metrics
# - Saves everything to /content/heart_eda_artifacts and zips it

import os, json, shutil
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, average_precision_score, brier_score_loss
from sklearn.feature_selection import mutual_info_classif

import inspect

from google.colab import drive
drive.mount('/content/drive')

# ------------ 0) Upload (or set CSV_PATH directly) ------------
CSV_PATH = "/content/drive/My Drive/Heart/heart_disease_uci.csv"
# if not CSV_PATH:
#     try:
#         from google.colab import files  # type: ignore
#         print("Upload your heart-disease CSV…")
#         uploaded = files.upload()
#         CSV_PATH = list(uploaded.keys())[0]
#     except Exception:
#         raise SystemExit("Please set CSV_PATH to your file or use Colab 'files.upload()'.")

# ------------ 1) Setup output dirs ------------
ARTIFACTS_DIR = Path("/content/drive/My Drive/Heart/heart_eda_artifacts")
PLOTS_DIR = ARTIFACTS_DIR / "plots"
ARTIFACTS_DIR.mkdir(parents=True, exist_ok=True)
PLOTS_DIR.mkdir(parents=True, exist_ok=True)

# ------------ 2) Load & light cleaning ------------
df = pd.read_csv(CSV_PATH)
raw = df.copy()

# Normalize column names to snake_case (non-destructive: keep original for reference)
df.columns = [str(c).strip().lower().replace(" ", "_") for c in df.columns]

# Replace common missing tokens in object cols
MISS_TOKENS = {"?", "na", "n/a", "none", ""}
for c in df.columns:
    if df[c].dtype == "object":
        df[c] = df[c].replace({t: np.nan for t in MISS_TOKENS})

# Try to coerce numeric-looking object columns
for c in df.columns:
    if df[c].dtype == "object":
        coerced = pd.to_numeric(df[c], errors="ignore")
        # promote only if conversion happened
        if not isinstance(coerced.dtype, pd.StringDtype) and not (coerced.dtype == object):
            df[c] = coerced

# Ensure a row id
if "row_id" not in df.columns:
    df.insert(0, "row_id", np.arange(1, len(df) + 1))

# ------------ 3) Target detection & binarization ------------
def detect_target(cols):
    cands = ["target", "num", "disease", "has_disease", "hd", "class", "label"]
    for k in cands:
        if k in cols: return k
    # heuristic: a single binary column
    bin_cols = []
    for c in cols:
        vals = pd.Series(df[c]).dropna().unique().tolist()
        try:
            sval = set(vals)
        except TypeError:
            continue
        if len(sval) == 2 and sval.issubset({0,1}):
            bin_cols.append(c)
    return bin_cols[0] if len(bin_cols) == 1 else None

cols = list(df.columns)
target_raw = detect_target(cols)
if target_raw is None:
    raise SystemExit(
        "Could not find a target column. Expected one of: "
        "['target','num','disease','has_disease','hd','class','label'] "
        "or a single binary 0/1 column."
    )

y_raw = df[target_raw]
if y_raw.dropna().nunique() > 2:
    df["has_disease"] = (y_raw > 0).astype(int)
    target = "has_disease"
    target_note = f"Binarized '{target_raw}' via (>0 -> 1)."
else:
    uniq = sorted(pd.Series(y_raw.dropna().unique()).tolist())
    if uniq == [0,1]:
        df["has_disease"] = y_raw.astype(int)
        target = "has_disease"
        target_note = f"Used '{target_raw}' as-is (0/1)."
    else:
        mapping = {uniq[0]:0, uniq[-1]:1}
        df["has_disease"] = y_raw.map(mapping).astype(int)
        target = "has_disease"
        target_note = f"Mapped '{target_raw}' with {mapping} -> 'has_disease'."

print("Target:", target, "|", target_note)

# ------------ 4) Recode physiologic invalids -> NaN ------------
invalid_rules = {
    "trestbps": lambda s: s <= 0,
    "chol":     lambda s: s <= 0,
    "oldpeak":  lambda s: s < 0
}
changed_counts = {}
for col, rule in invalid_rules.items():
    if col in df.columns:
        mask = rule(df[col]) & df[col].notna()
        changed_counts[col] = int(mask.sum())
        df.loc[mask, col] = np.nan
if changed_counts:
    print("Re-coded invalid → NaN:", changed_counts)

# ------------ 5) Data Dictionary ------------
def example_vals(series, maxn=8):
    vals = series.dropna().unique().tolist()
    if not vals: return ""
    return ", ".join(map(str, vals[:maxn])) + ("..." if len(vals) > maxn else "")

dq = pd.DataFrame({
    "column": df.columns,
    "dtype": [str(df[c].dtype) for c in df.columns],
    "missing_pct": df.isna().mean().round(4),
    "n_unique": df.nunique()
})
dq["example_values"] = [example_vals(df[c]) for c in df.columns]
dq = dq.sort_values(["missing_pct","column"], ascending=[False,True]).reset_index(drop=True)
dq_path = ARTIFACTS_DIR / "data_dictionary.csv"
dq.to_csv(dq_path, index=False)
display(dq.head(20))

# ------------ 6) Prevalence & slices ------------
y = df[target].astype(int)
prev = pd.DataFrame({
    "metric": ["prevalence","positives","negatives","total"],
    "value": [float(y.mean()), int(y.sum()), int((1-y).sum()), int(len(y))]
})
prev.to_csv(ARTIFACTS_DIR / "prevalence_summary.csv", index=False)
display(prev)

sex_col = None
for cand in ["sex","gender"]:
    if cand in df.columns:
        sex_col = cand; break
if sex_col:
    by_sex = df.groupby(sex_col)[target].mean().reset_index().rename(columns={target:"prevalence"})
    by_sex.to_csv(ARTIFACTS_DIR / "prevalence_by_sex.csv", index=False)
    display(by_sex)

age_col = "age" if "age" in df.columns else None
if age_col:
    bands = pd.cut(df[age_col], bins=[df[age_col].min()-1,45,65,df[age_col].max()+1], labels=["<45","45–64","65+"])
    by_age = pd.DataFrame({"age_band": bands, target: y}).groupby("age_band")[target].mean().reset_index().rename(columns={target:"prevalence"})
    by_age.to_csv(ARTIFACTS_DIR / "prevalence_by_age_band.csv", index=False)
    display(by_age)

# ------------ 7) Range checks ------------
expected_bounds = {
    "age":      (18, 100),
    "trestbps": (70, 220),
    "chol":     (80, 700),
    "thalach":  (60, 230),
    "oldpeak":  (0, 7),
}
range_rows = []
for c,(lo,hi) in expected_bounds.items():
    if c in df.columns:
        bad = (~df[c].between(lo,hi)) & df[c].notna()
        range_rows.append({"column": c, "out_of_range_count": int(bad.sum()), "lo": lo, "hi": hi})
range_df = pd.DataFrame(range_rows) if range_rows else pd.DataFrame(columns=["column","out_of_range_count","lo","hi"])
range_df.to_csv(ARTIFACTS_DIR / "range_checks.csv", index=False)
display(range_df)

# ------------ 8) Plots (matplotlib, one figure per plot) ------------
def hist_by_label(col):
    if col not in df.columns: return
    fig = plt.figure()
    neg = df[df[target]==0][col].dropna()
    pos = df[df[target]==1][col].dropna()
    plt.hist(neg, alpha=0.6, label="no disease")
    plt.hist(pos, alpha=0.6, label="has disease")
    plt.xlabel(col); plt.ylabel("count"); plt.title(f"Distribution of {col} by label"); plt.legend()
    plt.tight_layout(); plt.show()
    fig.savefig(PLOTS_DIR / f"hist_{col}.png", dpi=130, bbox_inches="tight")

for col in [c for c in ["age","trestbps","chol","thalach","oldpeak"] if c in df.columns]:
    hist_by_label(col)

def bar_rate(col):
    if col not in df.columns: return
    tab = df[[col, target]].dropna()
    if tab.empty: return
    rate = tab.groupby(col)[target].mean().sort_values(ascending=False)
    fig = plt.figure()
    plt.bar(rate.index.astype(str), rate.values)
    plt.xlabel(col); plt.ylabel("disease prevalence"); plt.title(f"Prevalence by {col}")
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout(); plt.show()
    fig.savefig(PLOTS_DIR / f"rate_{col}.png", dpi=130, bbox_inches="tight")

for col in ["cp","restecg","slope","thal","exang","ca","fbs","sex"]:
    if col in df.columns: bar_rate(col)

# ------------ 9) Feature typing for stats ------------
feature_cols = [c for c in df.columns if c not in {"row_id", target, target_raw}]
num_cols, cat_cols = [], []
for c in feature_cols:
    if pd.api.types.is_numeric_dtype(df[c]):
        if df[c].nunique(dropna=True) <= 10:
            cat_cols.append(c)
        else:
            num_cols.append(c)
    else:
        cat_cols.append(c)

# ------------ 10) Mutual Information (train split) ------------
X_train, X_test, y_train, y_test = train_test_split(
    df[feature_cols], y, test_size=0.30, random_state=42, stratify=y
)

# Simple encodings for MI:
num_imp = SimpleImputer(strategy="median")
cat_imp = SimpleImputer(strategy="most_frequent")
Xn = pd.DataFrame(num_imp.fit_transform(X_train[num_cols]), columns=num_cols, index=X_train.index) if num_cols else pd.DataFrame(index=X_train.index)
Xc_base = pd.DataFrame(cat_imp.fit_transform(X_train[cat_cols]), columns=cat_cols, index=X_train.index) if cat_cols else pd.DataFrame(index=X_train.index)
ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
if not Xc_base.empty:
    Xc = ohe.fit_transform(Xc_base)
    cat_feat_names = ohe.get_feature_names_out(cat_cols)
    Xc = pd.DataFrame(Xc, columns=cat_feat_names, index=X_train.index)
else:
    Xc = pd.DataFrame(index=X_train.index)
X_enc = pd.concat([Xn, Xc], axis=1)

discrete_mask = np.array([False]*len(num_cols) + [True]*len(Xc.columns))
mi = mutual_info_classif(X_enc.values, y_train.values, discrete_features=discrete_mask, random_state=42)

# Aggregate MI back to original features
mi_agg = {}
start = len(num_cols)
for i, name in enumerate(num_cols):
    mi_agg[name] = mi[i]
for j, name in enumerate(Xc.columns):
    orig = name.split("_")[0]
    mi_agg[orig] = mi_agg.get(orig, 0.0) + mi[start + j]
mi_df = pd.DataFrame({"feature": list(mi_agg.keys()), "mutual_information": list(mi_agg.values())}).sort_values("mutual_information", ascending=False)
mi_df.to_csv(ARTIFACTS_DIR / "mutual_information.csv", index=False)
display(mi_df.head(15))

# ------------ 11) Numeric correlation ------------
if num_cols:
    corr = X_train[num_cols].corr(method="pearson")
    corr.to_csv(ARTIFACTS_DIR / "numeric_correlation.csv")
    display(corr)

# ------------ 12) Quick baseline (holdout) ------------
num_tf = SimpleImputer(strategy="median", add_indicator=True)
cat_tf = Pipeline([
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("oh", OneHotEncoder(handle_unknown="ignore"))
])
ct = ColumnTransformer([
    ("num", num_tf, num_cols),
    ("cat", cat_tf, cat_cols)
])
clf = LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear")
pipe = Pipeline([("prep", ct), ("clf", clf)])
pipe.fit(X_train, y_train)
proba = pipe.predict_proba(X_test)[:,1]
auroc = roc_auc_score(y_test, proba)
ap = average_precision_score(y_test, proba)
brier = brier_score_loss(y_test, proba)

def expected_calibration_error(y_true, y_prob, n_bins=10):
    bins = np.linspace(0.0, 1.0, n_bins+1)
    idx = np.digitize(y_prob, bins) - 1
    ece = 0.0
    for b in range(n_bins):
        mask = idx == b
        if not np.any(mask): continue
        conf = y_prob[mask].mean()
        acc = y_true[mask].mean()
        ece += (np.sum(mask)/len(y_true)) * abs(acc - conf)
    return float(ece)

ece = expected_calibration_error(y_test.values, proba, n_bins=10)
metrics_df = pd.DataFrame({
    "metric": ["AUROC (holdout)", "PR AUC (holdout)", "Brier (holdout)", "ECE@10bins (holdout)"],
    "value": [round(auroc,4), round(ap,4), round(brier,4), round(ece,4)]
})
metrics_df.to_csv(ARTIFACTS_DIR / "baseline_metrics.csv", index=False)
display(metrics_df)

# ------------ 13) Go / No-Go summary ------------
go_df = pd.DataFrame({
    "check": ["Baseline AUROC >= 0.75", "Per-feature missingness <= 30%", "Sample size ≥ 1500 (nice-to-have)"],
    "pass":  [auroc >= 0.75, bool(dq["missing_pct"].max() <= 0.30), len(df) >= 1500]
})
go_df.to_csv(ARTIFACTS_DIR / "go_no_go.csv", index=False)
display(go_df)

# ------------ 14) Notes & bundle ------------
notes = {
    "target_note": target_note,
    "invalid_recode_counts": changed_counts,
    "csv_used": CSV_PATH,
}
with open(ARTIFACTS_DIR / "eda_notes.json", "w") as f:
    json.dump(notes, f, indent=2)

zip_path = shutil.make_archive("/content/heart_eda_artifacts", "zip", ARTIFACTS_DIR)
print("All artifacts saved to:", ARTIFACTS_DIR)
print("Zipped bundle:", zip_path)

# If desired in Colab, auto-download the zip:
# from google.colab import files; files.download(zip_path)
```

```python
# === Phase 3: Data Preparation & Training Recipe (Colab-ready) ===
# Leakage-safe preprocessing + model training with threshold tuning & calibration.
# Artifacts saved under /content/heart_phase3_artifacts

import os, json, shutil, inspect
from pathlib import Path
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import (
    roc_auc_score, average_precision_score, brier_score_loss,
    confusion_matrix, precision_recall_curve, roc_curve
)
import joblib

# ---------- 0) Load data ----------
CSV_PATH = "/content/drive/My Drive/Heart/heart_disease_uci.csv"
# if not CSV_PATH:
#     try:
#         from google.colab import files  # type: ignore
#         print("Upload your heart-disease CSV…")
#         uploaded = files.upload()
#         CSV_PATH = list(uploaded.keys())[0]
#     except Exception:
#         raise SystemExit("Please set CSV_PATH or use upload.")

df = pd.read_csv(CSV_PATH)
raw = df.copy()

# normalize column names a bit
df.columns = [str(c).strip().lower().replace(" ", "_") for c in df.columns]

# clean common missing tokens in object columns
MISS_TOKENS = {"?", "na", "n/a", "none", ""}
for c in df.columns:
    if df[c].dtype == "object":
        df[c] = df[c].replace({t: np.nan for t in MISS_TOKENS})

# coerce numeric-like object columns
for c in df.columns:
    if df[c].dtype == "object":
        coerced = pd.to_numeric(df[c], errors="ignore")
        if not (coerced.dtype == object):
            df[c] = coerced

if "row_id" not in df.columns:
    df.insert(0, "row_id", np.arange(1, len(df)+1))

# ---------- 1) Target detection & binarization ----------
def detect_target(cols):
    cands = ["target","num","disease","has_disease","hd","class","label"]
    for k in cands:
        if k in cols: return k
    # fallback: single binary 0/1 column
    bins = []
    for c in cols:
        vals = pd.Series(df[c]).dropna().unique().tolist()
        try:
            sval = set(vals)
        except TypeError:
            continue
        if len(sval) == 2 and sval.issubset({0,1}):
            bins.append(c)
    return bins[0] if len(bins) == 1 else None

target_raw = detect_target(list(df.columns))
if target_raw is None:
    raise SystemExit("No target found. Add a column named one of: target/num/disease/has_disease/hd/class/label or a single 0/1 column.")

y_raw = df[target_raw]
if y_raw.dropna().nunique() > 2:
    df["has_disease"] = (y_raw > 0).astype(int)
    target_note = f"Binarized '{target_raw}' via (>0 -> 1)."
else:
    uniq = sorted(pd.Series(y_raw.dropna().unique()).tolist())
    if uniq == [0,1]:
        df["has_disease"] = y_raw.astype(int)
        target_note = f"Used '{target_raw}' as-is (0/1)."
    else:
        mapping = {uniq[0]:0, uniq[-1]:1}
        df["has_disease"] = y_raw.map(mapping).astype(int)
        target_note = f"Mapped '{target_raw}' with {mapping} -> 'has_disease'."

TARGET = "has_disease"

# ---------- 2) Physiologic validity recode -> NaN ----------
invalid_rules = {
    "trestbps": lambda s: s <= 0,
    "chol":     lambda s: s <= 0,
    "oldpeak":  lambda s: s < 0
}
invalid_counts = {}
for col, rule in invalid_rules.items():
    if col in df.columns:
        m = rule(df[col]) & df[col].notna()
        invalid_counts[col] = int(m.sum())
        df.loc[m, col] = np.nan

# ---------- 3) Feature roles & variants ----------
ALL_FEATURES = [c for c in df.columns if c not in {"row_id", TARGET, target_raw}]
# heuristic: numeric vs categorical
num_cols, cat_cols = [], []
for c in ALL_FEATURES:
    if pd.api.types.is_numeric_dtype(df[c]):
        if df[c].nunique(dropna=True) <= 10:
            cat_cols.append(c)
        else:
            num_cols.append(c)
    else:
        cat_cols.append(c)

# pathway-like columns (often high-missing or encode workflow)
PATHWAY = [c for c in ["ca","thal","slope"] if c in df.columns]

VARIANTS = {
    "full": {
        "features": ALL_FEATURES.copy(),
    },
    "no_pathway": {
        "features": [c for c in ALL_FEATURES if c not in PATHWAY],
    },
}

# ---------- 4) Version-proof OneHotEncoder ----------
def make_ohe(dense=True, **kwargs):
    params = inspect.signature(OneHotEncoder).parameters
    if dense:
        if "sparse_output" in params:   # sklearn >= 1.2
            kwargs["sparse_output"] = False
        else:                           # sklearn < 1.2
            kwargs["sparse"] = False
    return OneHotEncoder(handle_unknown="ignore", **kwargs)

# ---------- 5) Train/Val/Test split (leakage safe) ----------
y = df[TARGET].astype(int).values
X_all = df.copy()

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X_all, y, test_size=0.20, random_state=42, stratify=y
)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=0.25, random_state=42, stratify=y_trainval
)
# (now: 60/20/20 split)

def build_preprocessor(X, feature_list):
    # Redo type detection on the subset to be precise
    num, cat = [], []
    for c in feature_list:
        if pd.api.types.is_numeric_dtype(X[c]):
            if X[c].nunique(dropna=True) <= 10:
                cat.append(c)
            else:
                num.append(c)
        else:
            cat.append(c)
    num_tf = SimpleImputer(strategy="median", add_indicator=True)
    cat_tf = Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("oh", make_ohe(dense=True))
    ])
    pre = ColumnTransformer([
        ("num", num_tf, num),
        ("cat", cat_tf, cat)
    ])
    return pre, num, cat

def fit_calibrated_logreg(Xtr, ytr, Xva, yva, features):
    pre, num, cat = build_preprocessor(Xtr, features)
    base = Pipeline([
        ("pre", pre),
        ("clf", LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear"))
    ])
    base.fit(Xtr[features], ytr)
    # isotonic calibration on validation (avoids leakage)
    calib = CalibratedClassifierCV(base, method="isotonic", cv="prefit")
    calib.fit(Xva[features], yva)
    return calib, {"num_cols": num, "cat_cols": cat}

def pick_threshold_for_sensitivity(y_true, y_prob, min_sens=0.90):
    # sweep unique probabilities (plus 1.0) to find highest specificity s.t. sens >= min_sens
    order = np.argsort(-y_prob)
    thresh_candidates = np.r_[np.unique(y_prob[order]), [0.0]]
    best = {"threshold": 0.5, "sensitivity": 0.0, "specificity": 0.0}
    P = (y_true == 1).sum()
    N = (y_true == 0).sum()
    for t in thresh_candidates:
        yhat = (y_prob >= t).astype(int)
        TP = int(((yhat == 1) & (y_true == 1)).sum())
        TN = int(((yhat == 0) & (y_true == 0)).sum())
        sens = TP / P if P else 0.0
        spec = TN / N if N else 0.0
        if sens >= min_sens:
            # maximize specificity, break ties with higher threshold to reduce FPs
            if (spec > best["specificity"]) or (spec == best["specificity"] and t > best["threshold"]):
                best = {"threshold": float(t), "sensitivity": float(sens), "specificity": float(spec)}
    # fallback to Youden J if constraint impossible
    if best["sensitivity"] < min_sens:
        fpr, tpr, thr = roc_curve(y_true, y_prob)
        j = tpr - fpr
        k = int(np.argmax(j))
        best = {"threshold": float(thr[k]), "sensitivity": float(tpr[k]), "specificity": float(1 - fpr[k])}
    return best

def evaluate(y_true, y_prob, threshold, groups: dict = None):
    yhat = (y_prob >= threshold).astype(int)
    TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
    metrics = {
        "AUROC": float(roc_auc_score(y_true, y_prob)),
        "AUPRC": float(average_precision_score(y_true, y_prob)),
        "Brier": float(brier_score_loss(y_true, y_prob)),
        "threshold": float(threshold),
        "TP": int(TP), "FP": int(FP), "TN": int(TN), "FN": int(FN),
        "Sensitivity": float(TP / (TP + FN) if (TP + FN) else 0.0),
        "Specificity": float(TN / (TN + FP) if (TN + FP) else 0.0),
        "PPV": float(TP / (TP + FP) if (TP + FP) else 0.0),
        "NPV": float(TN / (TN + FN) if (TN + FN) else 0.0),
        "Prevalence": float(np.mean(y_true))
    }
    # group slices
    slices = {}
    if groups:
        for name, g in groups.items():
            mask = pd.Series(g).notna().values
            # evaluate only where group value is not NaN
            y_t = y_true[mask]
            p_t = y_prob[mask]
            v_t = np.array(g)[mask]
            for lvl in pd.Series(v_t).unique():
                m = v_t == lvl
                if m.sum() == 0:
                    continue
                try:
                    sub = evaluate(y_t[m], p_t[m], threshold, groups=None)
                    slices[f"{name}={lvl}"] = {k: sub[k] for k in ["Sensitivity","Specificity","PPV","NPV","Prevalence","AUROC","AUPRC"]}
                except Exception:
                    pass
    return metrics, slices

ART_DIR = Path("/content/drive/My Drive/Heart/heart_phase3_artifacts")
ART_DIR.mkdir(parents=True, exist_ok=True)

results = {}
for variant_name, cfg in VARIANTS.items():
    feats = cfg["features"]
    # fit on Train; calibrate on Val; choose threshold on Val
    model, prep_info = fit_calibrated_logreg(X_train, y_train, X_val, y_val, feats)
    # choose threshold to meet Sens >= 0.90 on validation
    val_proba = model.predict_proba(X_val[feats])[:,1]
    th = pick_threshold_for_sensitivity(y_val, val_proba, min_sens=0.90)
    # evaluate on Test
    test_proba = model.predict_proba(X_test[feats])[:,1]
    sex_col = "sex" if "sex" in X_test.columns else None
    age_col = "age" if "age" in X_test.columns else None
    age_band = None
    if age_col:
        a = X_test[age_col]
        age_band = pd.cut(a, bins=[a.min()-1,45,65,a.max()+1], labels=["<45","45–64","65+"])
    groups = {}
    if sex_col: groups["sex"] = X_test[sex_col]
    if age_band is not None: groups["age_band"] = age_band

    overall, slices = evaluate(y_test, test_proba, th["threshold"], groups=groups)
    results[variant_name] = {
        "features_used": feats,
        "prep_info": prep_info,
        "threshold_choice_on_val": th,
        "metrics_test": overall,
        "slices_test": slices,
    }
    # save model + feature list + threshold
    vdir = ART_DIR / variant_name
    vdir.mkdir(exist_ok=True)
    joblib.dump(model, vdir / "calibrated_logreg_pipeline.joblib")
    with open(vdir / "features.json", "w") as f:
        json.dump({"features": feats, "num_cols": prep_info["num_cols"], "cat_cols": prep_info["cat_cols"]}, f, indent=2)
    with open(vdir / "threshold.json", "w") as f:
        json.dump(th, f, indent=2)
    with open(vdir / "metrics_test.json", "w") as f:
        json.dump(overall, f, indent=2)
    with open(vdir / "slices_test.json", "w") as f:
        json.dump(slices, f, indent=2)

# Save run manifest
manifest = {
    "csv_used": CSV_PATH,
    "n_rows": int(len(df)),
    "target_note": target_note,
    "invalid_recode_counts": invalid_counts,
    "pathway_columns_present": PATHWAY,
}
with open(ART_DIR / "run_manifest.json", "w") as f:
    json.dump(manifest, f, indent=2)

# also dump a compact summary table
def summarize(res):
    rows = []
    for name, r in res.items():
        m = r["metrics_test"]
        rows.append([name, m["AUROC"], m["AUPRC"], m["Brier"], m["Sensitivity"], m["Specificity"], m["PPV"], m["NPV"], m["Prevalence"], r["threshold_choice_on_val"]["threshold"]])
    return pd.DataFrame(rows, columns=["variant","AUROC","AUPRC","Brier","Sensitivity","Specificity","PPV","NPV","Prevalence","threshold"])

summary_df = summarize(results)
summary_df.to_csv(ART_DIR / "summary_metrics.csv", index=False)
display(summary_df)

print("Artifacts saved in:", ART_DIR)
print("Subfolders:", [p.name for p in ART_DIR.iterdir()])
# If you want a ZIP to download:
# import shutil; shutil.make_archive("/content/heart_phase3_artifacts", "zip", ART_DIR)
```

```python
# === CRISP-DM Phase 4: Modeling (Colab-ready) ==================================
# Models: Logistic Regression, RandomForest, XGBoost (if available)
# - Two feature variants: "full" and "no_pathway" (drops ca/thal/slope if present)
# - 60/20/20 Train/Val/Test split
# - RandomizedSearchCV (5-fold stratified) -> pick best by AUROC
# - Isotonic calibration on Validation (leakage-safe)
# - Thresholds:
#     A) Sensitivity >= 0.90 (maximize specificity subject to Sens constraint)
#     B) Cost-minimizing (CostFN=10, CostFP=2) on Validation
# - Test-set metrics overall + by sex/age slices
# - Artifacts saved to /content/heart_phase4_artifacts (pipelines, metrics, thresholds, summary)
# ================================================================================
import os, json, math, inspect, warnings, shutil
from pathlib import Path
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import (roc_auc_score, average_precision_score, brier_score_loss,
                             confusion_matrix, roc_curve)
from scipy.stats import loguniform
import joblib

# Optional: XGBoost (will be skipped if not installed)
try:
    import xgboost as xgb
    HAVE_XGB = True
except Exception:
    HAVE_XGB = False

warnings.filterwarnings("ignore")

# ------------------- Inputs -------------------
CSV_PATH = "/content/drive/My Drive/Heart/heart_disease_uci.csv"
# if not CSV_PATH:
#     try:
#         from google.colab import files  # type: ignore
#         print("Upload your heart-disease CSV…")
#         uploaded = files.upload()
#         CSV_PATH = list(uploaded.keys())[0]
#     except Exception:
#         raise SystemExit("Set CSV_PATH or upload the CSV in Colab.")

ART_DIR = Path("/content/drive/My Drive/Heart/heart_phase4_artifacts"); ART_DIR.mkdir(parents=True, exist_ok=True)

# ------------------- Load & normalize -------------------
df = pd.read_csv(CSV_PATH)
df.columns = [str(c).strip().lower().replace(" ", "_") for c in df.columns]

# Clean common missing tokens in object cols
MISS_TOKENS = {"?", "na", "n/a", "none", ""}
for c in df.columns:
    if df[c].dtype == "object":
        df[c] = df[c].replace({t: np.nan for t in MISS_TOKENS})

# Coerce numeric-like objects
for c in df.columns:
    if df[c].dtype == "object":
        coerced = pd.to_numeric(df[c], errors="ignore")
        if not (coerced.dtype == object):
            df[c] = coerced

# Row id (for reference only)
if "row_id" not in df.columns:
    df.insert(0, "row_id", np.arange(1, len(df)+1))

# ------------------- Target detection & binarization -------------------
def detect_target(cols):
    cands = ["target","num","disease","has_disease","hd","class","label"]
    for k in cands:
        if k in cols: return k
    # fallback: a single 0/1 column
    bin_cols = []
    for c in cols:
        vals = pd.Series(df[c]).dropna().unique().tolist()
        try:
            sval = set(vals)
        except TypeError:
            continue
        if len(sval) == 2 and sval.issubset({0,1}):
            bin_cols.append(c)
    return bin_cols[0] if len(bin_cols) == 1 else None

target_raw = detect_target(list(df.columns))
if target_raw is None:
    raise SystemExit("No target found. Expected one of target/num/disease/has_disease/hd/class/label or a single 0/1 column.")

y_raw = df[target_raw]
if y_raw.dropna().nunique() > 2:
    df["has_disease"] = (y_raw > 0).astype(int)
    target_note = f"Binarized '{target_raw}' via (>0 -> 1)."
else:
    uniq = sorted(pd.Series(y_raw.dropna().unique()).tolist())
    if uniq == [0,1]:
        df["has_disease"] = y_raw.astype(int)
        target_note = f"Used '{target_raw}' as-is (0/1)."
    else:
        mapping = {uniq[0]:0, uniq[-1]:1}
        df["has_disease"] = y_raw.map(mapping).astype(int)
        target_note = f"Mapped '{target_raw}' with {mapping} -> 'has_disease'."
TARGET = "has_disease"

# ------------------- Physiologic validity recode -> NaN -------------------
invalid_rules = {
    "trestbps": lambda s: s <= 0,
    "chol":     lambda s: s <= 0,
    "oldpeak":  lambda s: s < 0,
}
for col, rule in invalid_rules.items():
    if col in df.columns:
        df.loc[rule(df[col]) & df[col].notna(), col] = np.nan

# ------------------- Feature roles & variants -------------------
ALL_FEATURES = [c for c in df.columns if c not in {"row_id", TARGET, target_raw}]

PATHWAY = [c for c in ["ca","thal","slope"] if c in df.columns]  # often high-missing/workflow-encoded
VARIANTS = {
    "full": {"features": ALL_FEATURES.copy()},
    "no_pathway": {"features": [c for c in ALL_FEATURES if c not in PATHWAY]},
}

# ------------------- Version-proof OneHotEncoder -------------------
def make_ohe(dense=True, **kwargs):
    params = inspect.signature(OneHotEncoder).parameters
    if dense:
        if "sparse_output" in params:   # sklearn >= 1.2
            kwargs["sparse_output"] = False
        else:                           # sklearn < 1.2
            kwargs["sparse"] = False
    return OneHotEncoder(handle_unknown="ignore", **kwargs)

# ------------------- Split -------------------
y = df[TARGET].astype(int).values
X_all = df.copy()

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X_all, y, test_size=0.20, random_state=42, stratify=y
)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=0.25, random_state=42, stratify=y_trainval
)  # 60/20/20

# ------------------- Preprocessor builder -------------------
def build_preprocessor(X, feature_list):
    num, cat = [], []
    for c in feature_list:
        if pd.api.types.is_numeric_dtype(X[c]):
            if X[c].nunique(dropna=True) <= 10:
                cat.append(c)
            else:
                num.append(c)
        else:
            cat.append(c)
    num_tf = SimpleImputer(strategy="median", add_indicator=True)
    cat_tf = Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                       ("oh", make_ohe(dense=True))])
    pre = ColumnTransformer([("num", num_tf, num), ("cat", cat_tf, cat)])
    return pre, num, cat

# ------------------- Models & search spaces -------------------
def model_spaces():
    models = []
    # Logistic Regression
    logreg = LogisticRegression(max_iter=1000, class_weight="balanced", solver="liblinear")
    log_space = {"clf__C": loguniform(1e-3, 1e2)}
    models.append(("logreg", logreg, log_space))

    # Random Forest
    rf = RandomForestClassifier(class_weight="balanced", n_estimators=500, random_state=42, n_jobs=-1)
    rf_space = {
        "clf__n_estimators": [300, 500, 700, 900],
        "clf__max_depth": [None, 4, 6, 8, 12],
        "clf__min_samples_leaf": [1, 2, 4, 8],
        "clf__max_features": ["sqrt", "log2", None],
    }
    models.append(("rf", rf, rf_space))

    # XGBoost (optional)
    if HAVE_XGB:
        xgb_model = xgb.XGBClassifier(
            objective="binary:logistic",
            eval_metric="logloss",
            tree_method="hist",
            n_estimators=600,
            learning_rate=0.05,
            subsample=0.9,
            colsample_bytree=0.9,
            reg_lambda=1.0,
            random_state=42,
            n_jobs=-1,
        )
        xgb_space = {
            "clf__max_depth": [3, 4, 5, 6],
            "clf__min_child_weight": [1, 2, 5],
            "clf__gamma": [0, 0.1, 0.2],
            "clf__reg_lambda": [0.5, 1.0, 2.0],
        }
        models.append(("xgb", xgb_model, xgb_space))
    return models

# ------------------- CV search helper -------------------
def cv_search(X, y, features, model_name, estimator, param_space, n_iter=20):
    pre, num, cat = build_preprocessor(X, features)
    pipe = Pipeline([("pre", pre), ("clf", estimator)])
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    search = RandomizedSearchCV(
        pipe, param_distributions=param_space, n_iter=n_iter, scoring="roc_auc",
        cv=cv, random_state=42, n_jobs=-1, verbose=0, refit=True
    )
    search.fit(X[features], y)
    return search, {"num_cols": num, "cat_cols": cat}

# ------------------- Thresholding helpers -------------------
def pick_threshold_for_sensitivity(y_true, y_prob, min_sens=0.90):
    order = np.argsort(-y_prob)
    thresh_candidates = np.r_[np.unique(y_prob[order]), [0.0]]
    best = {"threshold": 0.5, "sensitivity": 0.0, "specificity": 0.0}
    P = (y_true == 1).sum(); N = (y_true == 0).sum()
    for t in thresh_candidates:
        yhat = (y_prob >= t).astype(int)
        TP = int(((yhat == 1) & (y_true == 1)).sum())
        TN = int(((yhat == 0) & (y_true == 0)).sum())
        sens = TP / P if P else 0.0
        spec = TN / N if N else 0.0
        if sens >= min_sens:
            if (spec > best["specificity"]) or (spec == best["specificity"] and t > best["threshold"]):
                best = {"threshold": float(t), "sensitivity": float(sens), "specificity": float(spec)}
    # fallback to Youden J if needed
    if best["sensitivity"] < min_sens:
        fpr, tpr, thr = roc_curve(y_true, y_prob)
        j = tpr - fpr
        k = int(np.argmax(j))
        best = {"threshold": float(thr[k]), "sensitivity": float(tpr[k]), "specificity": float(1 - fpr[k])}
    return best

def pick_threshold_cost_min(y_true, y_prob, cost_fn=10.0, cost_fp=2.0):
    # minimize expected cost on validation
    thr = np.r_[np.unique(y_prob), [0.0]]
    P = (y_true == 1).sum(); N = (y_true == 0).sum()
    best = {"threshold":0.5, "expected_cost": math.inf, "Sensitivity":0.0, "Specificity":0.0}
    for t in thr:
        yhat = (y_prob >= t).astype(int)
        TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
        exp_cost = cost_fn*FN + cost_fp*FP
        sens = TP / (TP + FN) if (TP + FN) else 0.0
        spec = TN / (TN + FP) if (TN + FP) else 0.0
        if (exp_cost < best["expected_cost"]) or (exp_cost == best["expected_cost"] and t > best["threshold"]):
            best = {"threshold": float(t), "expected_cost": float(exp_cost),
                    "Sensitivity": float(sens), "Specificity": float(spec)}
    return best

def evaluate(y_true, y_prob, threshold, groups: dict = None):
    yhat = (y_prob >= threshold).astype(int)
    TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
    metrics = {
        "AUROC": float(roc_auc_score(y_true, y_prob)),
        "AUPRC": float(average_precision_score(y_true, y_prob)),
        "Brier": float(brier_score_loss(y_true, y_prob)),
        "threshold": float(threshold),
        "TP": int(TP), "FP": int(FP), "TN": int(TN), "FN": int(FN),
        "Sensitivity": float(TP / (TP + FN) if (TP + FN) else 0.0),
        "Specificity": float(TN / (TN + FP) if (TN + FP) else 0.0),
        "PPV": float(TP / (TP + FP) if (TP + FP) else 0.0),
        "NPV": float(TN / (TN + FN) if (TN + FN) else 0.0),
        "Prevalence": float(np.mean(y_true))
    }
    slices = {}
    if groups:
        for name, g in groups.items():
            mask = pd.Series(g).notna().values
            y_t = y_true[mask]; p_t = y_prob[mask]; v_t = np.array(g)[mask]
            for lvl in pd.Series(v_t).unique():
                m = v_t == lvl
                if m.sum() == 0: continue
                sub = evaluate(y_t[m], p_t[m], threshold, groups=None)[0]
                slices[f"{name}={lvl}"] = {k: sub[k] for k in
                    ["Sensitivity","Specificity","PPV","NPV","Prevalence","AUROC","AUPRC"]}
    return metrics, slices

# ------------------- Train, calibrate, evaluate across variants/models -------------------
summary_rows = []
for variant_name, cfg in VARIANTS.items():
    vdir = ART_DIR / variant_name; vdir.mkdir(exist_ok=True)
    feats = cfg["features"]

    best_model_name = None
    best_search = None
    best_auc = -1.0
    prep_info_best = None

    for model_name, est, space in model_spaces():
        search, prep_info = cv_search(X_train, y_train, feats, model_name, est, space, n_iter=20)
        mean_auc = search.best_score_
        # track best by CV AUROC
        if mean_auc > best_auc:
            best_auc = mean_auc
            best_model_name = model_name
            best_search = search
            prep_info_best = prep_info

        # save each model's CV result quickly
        (vdir / model_name).mkdir(exist_ok=True)
        with open(vdir / model_name / "cv_summary.json", "w") as f:
            json.dump({
                "best_params": search.best_params_,
                "best_cv_auroc": float(mean_auc)
            }, f, indent=2)

    # Calibrate the best on Validation
    best_pipe = best_search.best_estimator_
    calibrated = CalibratedClassifierCV(best_pipe, method="isotonic", cv="prefit")
    calibrated.fit(X_val[feats], y_val)

    # Thresholds on Validation
    val_proba = calibrated.predict_proba(X_val[feats])[:,1]
    thr_sens = pick_threshold_for_sensitivity(y_val, val_proba, min_sens=0.90)
    thr_cost = pick_threshold_cost_min(y_val, val_proba, cost_fn=10.0, cost_fp=2.0)

    # Evaluate on Test
    test_proba = calibrated.predict_proba(X_test[feats])[:,1]
    # groups
    sex_col = "sex" if "sex" in X_test.columns else None
    age_col = "age" if "age" in X_test.columns else None
    age_band = None
    if age_col:
        a = X_test[age_col]
        age_band = pd.cut(a, bins=[a.min()-1,45,65,a.max()+1], labels=["<45","45–64","65+"])
    groups = {}
    if sex_col: groups["sex"] = X_test[sex_col]
    if age_band is not None: groups["age_band"] = age_band

    overall_sens, slices_sens = evaluate(y_test, test_proba, thr_sens["threshold"], groups=groups)
    overall_cost, slices_cost = evaluate(y_test, test_proba, thr_cost["threshold"], groups=groups)

    # Save artifacts
    out = vdir / best_model_name
    out.mkdir(exist_ok=True)
    joblib.dump(calibrated, out / "calibrated_pipeline.joblib")
    with open(out / "features.json", "w") as f:
        json.dump({"features": feats, **prep_info_best}, f, indent=2)
    with open(out / "threshold_sensitivity.json", "w") as f:
        json.dump(thr_sens, f, indent=2)
    with open(out / "threshold_cost_min.json", "w") as f:
        json.dump(thr_cost, f, indent=2)
    with open(out / "metrics_test_at_sensitivity.json", "w") as f:
        json.dump(overall_sens, f, indent=2)
    with open(out / "metrics_test_at_costmin.json", "w") as f:
        json.dump(overall_cost, f, indent=2)
    with open(out / "slices_test_at_sensitivity.json", "w") as f:
        json.dump(slices_sens, f, indent=2)
    with open(out / "slices_test_at_costmin.json", "w") as f:
        json.dump(slices_cost, f, indent=2)

    # Add to run summary
    summary_rows.append([
        variant_name, best_model_name, best_auc,
        overall_sens["AUROC"], overall_sens["AUPRC"], overall_sens["Brier"],
        overall_sens["Sensitivity"], overall_sens["Specificity"], overall_sens["PPV"], overall_sens["NPV"],
        thr_sens["threshold"], "Sens>=0.90"
    ])
    summary_rows.append([
        variant_name, best_model_name, best_auc,
        overall_cost["AUROC"], overall_cost["AUPRC"], overall_cost["Brier"],
        overall_cost["Sensitivity"], overall_cost["Specificity"], overall_cost["PPV"], overall_cost["NPV"],
        thr_cost["threshold"], "CostMin(FN=10,FP=2)"
    ])

# Save summary
summary = pd.DataFrame(summary_rows, columns=[
    "variant","best_model_by_cv","best_cv_auroc",
    "test_AUROC","test_AUPRC","test_Brier",
    "test_Sensitivity","test_Specificity","test_PPV","test_NPV",
    "threshold","operating_point"
])
summary = summary.sort_values(["variant","operating_point"]).reset_index(drop=True)
summary_path = ART_DIR / "summary_metrics.csv"
summary.to_csv(summary_path, index=False)

# Save manifest
with open(ART_DIR / "manifest.json", "w") as f:
    json.dump({
        "csv_used": CSV_PATH,
        "n_rows": int(len(df)),
        "target_note": target_note,
        "pathway_columns_present": PATHWAY,
        "models_tried": ["logreg","rf"] + (["xgb"] if HAVE_XGB else []),
        "split": {"train": int(len(X_train)), "val": int(len(X_val)), "test": int(len(X_test))}
    }, f, indent=2)

from IPython.display import display
display(summary)
print("Artifacts saved to:", ART_DIR)
print("Subdirs:", [p.name for p in ART_DIR.iterdir()])
# Zip for download (uncomment to auto-download in Colab)
# zip_path = shutil.make_archive("/content/heart_phase4_artifacts", "zip", ART_DIR)
# from google.colab import files; files.download(zip_path)
```

```python
# === CRISP-DM Phase 5: Evaluation & Model Card (Colab-ready) =====================
# - Loads best model from /content/heart_phase4_artifacts (if there)
# - Recreates the 60/20/20 split (random_state=42) to evaluate on TEST
# - Computes ROC/PR, calibration (Brier, ECE, slope/intercept), threshold metrics
# - Decision Curve Analysis (net benefit) at both operating points
# - Fairness slices (sex, age bands) with parity gaps
# - Bootstrap 95% CIs for AUROC, AUPRC, Sens, Spec
# - Writes a ready-to-share Model Card markdown and zips artifacts
# ================================================================================

import os, json, math, shutil, inspect, textwrap
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    roc_auc_score, average_precision_score, brier_score_loss,
    confusion_matrix, roc_curve, precision_recall_curve
)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV

ART4 = Path("/content/drive/My Drive/Heart/heart_phase4_artifacts")
ART5 = Path("/content/drive/My Drive/Heart/heart_phase5_evaluation"); ART5.mkdir(parents=True, exist_ok=True)
PLOTS = ART5 / "plots"; PLOTS.mkdir(exist_ok=True)

# -------------------- Helpers --------------------
def make_ohe(dense=True, **kwargs):
    from sklearn.preprocessing import OneHotEncoder
    params = inspect.signature(OneHotEncoder).parameters
    if dense:
        if "sparse_output" in params: kwargs["sparse_output"] = False
        else: kwargs["sparse"] = False
    return OneHotEncoder(handle_unknown="ignore", **kwargs)

def expected_calibration_error(y_true, y_prob, n_bins=10):
    bins = np.linspace(0, 1, n_bins+1)
    idx = np.digitize(y_prob, bins) - 1
    ece = 0.0
    for b in range(n_bins):
        m = (idx == b)
        if not np.any(m): continue
        conf = y_prob[m].mean()
        acc = y_true[m].mean()
        ece += (m.mean()) * abs(acc - conf)
    return float(ece)

def calib_slope_intercept(y_true, y_prob, eps=1e-12):
    # Logistic calibration fit on test (for reporting only; not changing model)
    p = np.clip(y_prob, eps, 1-eps)
    logit = np.log(p/(1-p))
    X = np.column_stack([np.ones_like(logit), logit])
    # Weighted least squares ~ normal equations
    beta = np.linalg.pinv(X) @ y_true
    intercept, slope = beta[0], beta[1]
    return float(slope), float(intercept)

def decision_curve(y_true, y_prob, thresholds):
    # Net benefit: TP/N - FP/N * (pt/(1-pt))
    N = len(y_true); res = []
    for pt in thresholds:
        yhat = (y_prob >= pt).astype(int)
        TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
        nb = (TP/N) - (FP/N) * (pt/(1-pt))
        res.append((pt, nb))
    return pd.DataFrame(res, columns=["threshold","net_benefit"])

def bootstrap_ci(metric_fn, y_true, y_prob, B=300, seed=42):
    rng = np.random.default_rng(seed)
    vals = []
    n = len(y_true)
    for _ in range(B):
        idx = rng.integers(0, n, size=n)
        vals.append(metric_fn(y_true[idx], y_prob[idx]))
    arr = np.array(vals)
    return float(np.percentile(arr, 2.5)), float(np.percentile(arr, 97.5))

def metrics_at_threshold(y_true, y_prob, thr):
    yhat = (y_prob >= thr).astype(int)
    TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
    sens = TP/(TP+FN) if (TP+FN) else 0.0
    spec = TN/(TN+FP) if (TN+FP) else 0.0
    ppv  = TP/(TP+FP) if (TP+FP) else 0.0
    npv  = TN/(TN+FN) if (TN+FN) else 0.0
    return dict(TP=int(TP), FP=int(FP), TN=int(TN), FN=int(FN),
                Sensitivity=float(sens), Specificity=float(spec),
                PPV=float(ppv), NPV=float(npv))

# -------------------- Load data --------------------
CSV_PATH = "/content/drive/My Drive/Heart/heart_disease_uci.csv"
# if not CSV_PATH:
#     try:
#         from google.colab import files  # type: ignore
#         print("Upload your heart-disease CSV…")
#         uploaded = files.upload()
#         CSV_PATH = list(uploaded.keys())[0]
#     except Exception:
#         raise SystemExit("Set CSV_PATH or upload.")

df = pd.read_csv(CSV_PATH)
df.columns = [str(c).strip().lower().replace(" ", "_") for c in df.columns]

# Target detection
def detect_target(cols):
    for k in ["target","num","disease","has_disease","hd","class","label"]:
        if k in cols: return k
    # fallback: single binary column
    bin_cols = []
    for c in cols:
        vals = pd.Series(df[c]).dropna().unique().tolist()
        try: sval = set(vals)
        except TypeError: continue
        if len(sval) == 2 and sval.issubset({0,1}):
            bin_cols.append(c)
    return bin_cols[0] if len(bin_cols)==1 else None

target_raw = detect_target(list(df.columns))
y_raw = df[target_raw]
if y_raw.dropna().nunique() > 2:
    df["has_disease"] = (y_raw > 0).astype(int)
else:
    uniq = sorted(pd.Series(y_raw.dropna().unique()).tolist())
    if uniq == [0,1]:
        df["has_disease"] = y_raw.astype(int)
    else:
        df["has_disease"] = y_raw.map({uniq[0]:0, uniq[-1]:1}).astype(int)
TARGET = "has_disease"

# Physiologic validity recode -> NaN
for col, rule in {"trestbps":lambda s:s<=0, "chol":lambda s:s<=0, "oldpeak":lambda s:s<0}.items():
    if col in df.columns:
        df.loc[rule(df[col]) & df[col].notna(), col] = np.nan

# -------------------- Recreate 60/20/20 split --------------------
y = df[TARGET].astype(int).values
X_all = df.copy()
X_trainval, X_test, y_trainval, y_test = train_test_split(X_all, y, test_size=0.20, random_state=42, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.25, random_state=42, stratify=y_trainval)

# -------------------- Choose model to evaluate --------------------
# Prefer Phase-4 best from summary; otherwise fall back to calibrated logistic (no_pathway)
chosen = None
if ART4.exists():
    summ = ART4 / "summary_metrics.csv"
    if summ.exists():
        sm = pd.read_csv(summ)
        # pick the row with highest test_AUROC under Sens>=0.90 operating point
        cand = sm[sm["operating_point"]=="Sens>=0.90"].sort_values("test_AUROC", ascending=False).head(1)
        if len(cand):
            variant = cand.iloc[0]["variant"]
            best_model = cand.iloc[0]["best_model_by_cv"]
            chosen = {"variant": variant, "model": best_model}
    # Fallback: prefer no_pathway/logreg if present
    if chosen is None:
        for v in ["no_pathway","full"]:
            for m in ["logreg","rf","xgb"]:
                if (ART4 / v / m / "calibrated_pipeline.joblib").exists():
                    chosen = {"variant": v, "model": m}; break
            if chosen: break

# Load calibrated pipeline & thresholds if available
calibrated = None
features = None
thr_sens = None
thr_cost = None

if chosen is not None:
    vdir = ART4 / chosen["variant"] / chosen["model"]
    import joblib
    calibrated = joblib.load(vdir / "calibrated_pipeline.joblib")
    with open(vdir / "features.json") as f:
        meta = json.load(f)
    features = meta["features"]
    # thresholds (if saved)
    p_sens = vdir / "threshold_sensitivity.json"
    p_cost = vdir / "threshold_cost_min.json"
    if p_sens.exists():
        with open(p_sens) as f: thr_sens = json.load(f)
    if p_cost.exists():
        with open(p_cost) as f: thr_cost = json.load(f)

# If no Phase-4 model found, fit calibrated logistic (no_pathway) as fallback
if calibrated is None:
    ALL_FEATURES = [c for c in df.columns if c not in {"row_id", TARGET, target_raw}]
    PATHWAY = [c for c in ["ca","thal","slope"] if c in df.columns]
    features = [c for c in ALL_FEATURES if c not in PATHWAY]
    # type split
    num, cat = [], []
    for c in features:
        if pd.api.types.is_numeric_dtype(X_train[c]):
            if X_train[c].nunique(dropna=True) <= 10: cat.append(c)
            else: num.append(c)
        else:
            cat.append(c)
    pre = ColumnTransformer([
        ("num", SimpleImputer(strategy="median", add_indicator=True), num),
        ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                          ("oh", make_ohe(dense=True))]), cat)
    ])
    base = Pipeline([("pre", pre), ("clf", LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear"))])
    base.fit(X_train[features], y_train)
    calibrated = CalibratedClassifierCV(base, method="isotonic", cv="prefit")
    calibrated.fit(X_val[features], y_val)

# -------------------- Predict on TEST --------------------
proba = calibrated.predict_proba(X_test[features])[:,1]
auroc = roc_auc_score(y_test, proba)
auprc = average_precision_score(y_test, proba)
brier = brier_score_loss(y_test, proba)
ece = expected_calibration_error(y_test, proba, n_bins=10)
slope, intercept = calib_slope_intercept(y_test, proba)

# ROC / PR plots
fpr, tpr, roc_thr = roc_curve(y_test, proba)
prec, rec, pr_thr = precision_recall_curve(y_test, proba)

plt.figure(); plt.plot(fpr, tpr); plt.plot([0,1],[0,1],'--'); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title(f"ROC (AUC={auroc:.3f})"); plt.tight_layout(); plt.savefig(PLOTS/"roc.png", dpi=130); plt.show()
plt.figure(); plt.plot(rec, prec); plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title(f"PR (AUC={auprc:.3f})"); plt.tight_layout(); plt.savefig(PLOTS/"pr.png", dpi=130); plt.show()

# Reliability diagram
bins = np.linspace(0,1,11)
bin_ids = np.digitize(proba, bins)-1
bin_acc = [y_test[bin_ids==i].mean() if np.any(bin_ids==i) else np.nan for i in range(10)]
bin_conf = [np.nanmean(proba[bin_ids==i]) if np.any(bin_ids==i) else np.nan for i in range(10)]
plt.figure()
plt.plot([0,1],[0,1],'--')
plt.plot(bin_conf, bin_acc, marker='o')
plt.xlabel("Predicted prob"); plt.ylabel("Observed freq"); plt.title(f"Reliability (Brier={brier:.3f}, ECE={ece:.3f})")
plt.tight_layout(); plt.savefig(PLOTS/"reliability.png", dpi=130); plt.show()

# Thresholds: use saved ones if present; else compute equivalents on VAL
def pick_threshold_for_sensitivity(y_true, y_prob, min_sens=0.90):
    order = np.argsort(-y_prob); thr = np.r_[np.unique(y_prob[order]), [0.0]]
    best = {"threshold":0.5, "sensitivity":0.0, "specificity":0.0}
    P = (y_true==1).sum(); N = (y_true==0).sum()
    for t in thr:
        yhat = (y_prob>=t).astype(int)
        TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
        sens = TP/(TP+FN) if (TP+FN) else 0.0
        spec = TN/(TN+FP) if (TN+FP) else 0.0
        if sens>=min_sens and (spec>best["specificity"] or (spec==best["specificity"] and t>best["threshold"])):
            best = {"threshold":float(t), "sensitivity":float(sens), "specificity":float(spec)}
    return best

def pick_threshold_cost_min(y_true, y_prob, cost_fn=10.0, cost_fp=2.0):
    thr = np.r_[np.unique(y_prob), [0.0]]
    best = {"threshold":0.5, "expected_cost":math.inf}
    for t in thr:
        yhat=(y_prob>=t).astype(int)
        TN, FP, FN, TP = confusion_matrix(y_true, yhat, labels=[0,1]).ravel()
        cost = cost_fn*FN + cost_fp*FP
        if cost<best["expected_cost"] or (cost==best["expected_cost"] and t>best["threshold"]):
            best={"threshold":float(t),"expected_cost":float(cost)}
    return best

if thr_sens is None or thr_cost is None:
    # derive from validation
    val_proba = calibrated.predict_proba(X_val[features])[:,1]
    thr_sens = pick_threshold_for_sensitivity(y_val, val_proba, 0.90)
    thr_cost = pick_threshold_cost_min(y_val, val_proba, 10.0, 2.0)

# Metrics @ chosen thresholds on TEST
m_sens = metrics_at_threshold(y_test, proba, thr_sens["threshold"])
m_cost = metrics_at_threshold(y_test, proba, thr_cost["threshold"])

# Decision curves
pts = np.linspace(0.01, 0.99, 99)
dca = decision_curve(y_test, proba, pts)
plt.figure(); plt.plot(dca["threshold"], dca["net_benefit"])
plt.xlabel("Threshold probability"); plt.ylabel("Net benefit"); plt.title("Decision Curve (overall)")
plt.tight_layout(); plt.savefig(PLOTS/"decision_curve.png", dpi=130); plt.show()

# Fairness slices (sex, age bands)
# Align arrays to X_test's index to avoid label/position mismatch
y_test_s = pd.Series(y_test, index=X_test.index, name="y")
proba_s  = pd.Series(proba,  index=X_test.index, name="proba")

slices = {}

# Sex slices
if "sex" in X_test.columns:
    for lvl, grp in X_test.groupby("sex"):
        idx = grp.index
        yt = y_test_s.loc[idx].values
        pp = proba_s.loc[idx].values
        slices[f"sex={lvl}"] = {
            **metrics_at_threshold(yt, pp, thr_sens["threshold"]),
            "AUROC": float(roc_auc_score(yt, pp)),
            "AUPRC": float(average_precision_score(yt, pp)),
            "Prevalence": float(yt.mean())
        }

# Age-band slices
if "age" in X_test.columns:
    bands = pd.cut(X_test["age"],
                   bins=[X_test["age"].min()-1, 45, 65, X_test["age"].max()+1],
                   labels=["<45","45–64","65+"])
    for lvl in bands.cat.categories:
        idx = X_test.index[bands == lvl]
        if len(idx) == 0:
            continue
        yt = y_test_s.loc[idx].values
        pp = proba_s.loc[idx].values
        slices[f"age={lvl}"] = {
            **metrics_at_threshold(yt, pp, thr_sens["threshold"]),
            "AUROC": float(roc_auc_score(yt, pp)),
            "AUPRC": float(average_precision_score(yt, pp)),
            "Prevalence": float(yt.mean())
        }

# Bootstrap CIs
auroc_ci = bootstrap_ci(lambda yt, yp: roc_auc_score(yt, yp), y_test, proba)
auprc_ci = bootstrap_ci(lambda yt, yp: average_precision_score(yt, yp), y_test, proba)
sens_ci  = bootstrap_ci(lambda yt, yp: metrics_at_threshold(yt, yp, thr_sens["threshold"])["Sensitivity"], y_test, proba)
spec_ci  = bootstrap_ci(lambda yt, yp: metrics_at_threshold(yt, yp, thr_sens["threshold"])["Specificity"], y_test, proba)

# Save tabular outputs
summary = {
    "overall": {
        "AUROC": float(auroc), "AUROC_CI": auroc_ci,
        "AUPRC": float(auprc), "AUPRC_CI": auprc_ci,
        "Brier": float(brier), "ECE": float(ece),
        "Calib_slope": float(slope), "Calib_intercept": float(intercept),
    },
    "thresholds": {
        "sensitivity_first": {"threshold": float(thr_sens["threshold"]), **m_sens, "Sensitivity_CI": sens_ci, "Specificity_CI": spec_ci},
        "cost_min": {"threshold": float(thr_cost["threshold"]), **m_cost}
    },
    "fairness_slices_at_sens": slices,
    "features_used": features
}
with open(ART5/"evaluation_summary.json","w") as f: json.dump(summary, f, indent=2)

# CSV metrics export
pd.DataFrame([summary["overall"]]).to_csv(ART5/"overall_metrics.csv", index=False)
sens_row = pd.DataFrame([{"operating_point":"Sens>=0.90", **summary["thresholds"]["sensitivity_first"]}])
cost_row = pd.DataFrame([{"operating_point":"CostMin(FN=10,FP=2)", **summary["thresholds"]["cost_min"]}])
pd.concat([sens_row, cost_row], ignore_index=True).to_csv(ART5/"operating_points.csv", index=False)
pd.DataFrame.from_dict(summary["fairness_slices_at_sens"], orient="index").to_csv(ART5/"fairness_slices_at_sens.csv")

# -------------------- Model Card (markdown) --------------------
def md(x): return textwrap.dedent(x).strip()+"\n"
card = []
card.append(md(f"""
# Model Card — Heart Disease Binary Classifier (Phase 5 Evaluation)

**Intended use:** clinical decision support to prioritize further testing/referral.
**Not for** stand-alone diagnosis.

## Data & Split
- Dataset CSV: `{Path(CSV_PATH).name}`
- Split: 60/20/20 (random_state=42), evaluation on **TEST** set
- Features used: {', '.join(summary['features_used'])[:400]}{'...' if len(', '.join(summary['features_used']))>400 else ''}

## Overall Performance
- **AUROC:** {summary['overall']['AUROC']:.3f}  (95% CI: {summary['overall']['AUROC_CI'][0]:.3f}–{summary['overall']['AUROC_CI'][1]:.3f})
- **PR AUC:** {summary['overall']['AUPRC']:.3f}  (95% CI: {summary['overall']['AUPRC_CI'][0]:.3f}–{summary['overall']['AUPRC_CI'][1]:.3f})
- **Brier:** {summary['overall']['Brier']:.3f} | **ECE:** {summary['overall']['ECE']:.3f}
- **Calibration:** slope={summary['overall']['Calib_slope']:.3f}, intercept={summary['overall']['Calib_intercept']:.3f}

## Operating Points
**Sensitivity-first (>=0.90)**
- threshold={summary['thresholds']['sensitivity_first']['threshold']:.3f}
- Sens={summary['thresholds']['sensitivity_first']['Sensitivity']:.3f} (95% CI {summary['thresholds']['sensitivity_first']['Sensitivity_CI'][0]:.3f}–{summary['thresholds']['sensitivity_first']['Sensitivity_CI'][1]:.3f})
- Spec={summary['thresholds']['sensitivity_first']['Specificity']:.3f} (95% CI {summary['thresholds']['sensitivity_first']['Specificity_CI'][0]:.3f}–{summary['thresholds']['sensitivity_first']['Specificity_CI'][1]:.3f})
- PPV={summary['thresholds']['sensitivity_first']['PPV']:.3f} | NPV={summary['thresholds']['sensitivity_first']['NPV']:.3f}

**Cost-minimizing (FN=10, FP=2)**
- threshold={summary['thresholds']['cost_min']['threshold']:.3f}
- Sens={summary['thresholds']['cost_min']['Sensitivity']:.3f} | Spec={summary['thresholds']['cost_min']['Specificity']:.3f}
- PPV={summary['thresholds']['cost_min']['PPV']:.3f} | NPV={summary['thresholds']['cost_min']['NPV']:.3f}

## Fairness (Test @ Sens>=0.90 threshold)
"""))
# add fairness table
fair_df = pd.DataFrame.from_dict(summary["fairness_slices_at_sens"], orient="index")
card.append(fair_df.to_markdown() + "\n")
card.append(md("""
## Decision Curve
Net benefit curve saved in `plots/decision_curve.png`. Interpret relative to "treat-all" and "treat-none" lines (not shown by default).

## Known Risks & Mitigations
- **False negatives** are high-harm → we operate at Sens >= 0.90.
- **Pathway bias** (if using ca/thal/slope) → maintain a `no_pathway` variant and monitor drift/missingness.
- **Group gaps** → monitor Sens/Spec/PPV/NPV by sex and age bands in production; consider group-aware thresholds with clinical approval.

## Approval Gate (from Phase 1)
- Meets AUROC target (>=0.88), strong calibration, sensitivity-first operating point documented.
- Next: sign-off with clinical & compliance stakeholders.
"""))
with open(ART5/"model_card.md","w") as f: f.writelines(card)

# Save plots listing
for p in ["roc.png","pr.png","reliability.png","decision_curve.png"]:
    print("Saved", p, "->", PLOTS/p)

# Zip for download
zip_path = shutil.make_archive("/content/heart_phase5_evaluation", "zip", ART5)
print("Artifacts in:", ART5)
print("Zipped bundle:", zip_path)
```

```python
%%writefile "/content/drive/My Drive/Heart/serve/app.py"
# app.py — FastAPI inference service for Heart Disease classifier
import json, uvicorn, numpy as np, pandas as pd
from pathlib import Path
from fastapi import FastAPI
from pydantic import BaseModel, Field
import joblib
from typing import Optional, List, Dict, Any

# --- API & Schemas ---
app = FastAPI(
    title="Heart Disease Risk API",
    version="1.0",
    description="Decision support only. Not a stand-alone diagnostic.",
    openapi_url="/openapi.json",
    docs_url="/docs",
    redoc_url="/redoc",
)

# --- Load model + config ---
MODEL_DIR = Path(__file__).resolve().parent / "model"
PIPE = None
CFG  = {}
THRESH = 0.5

# expected feature list (if saved by Phase 4)
FEATURES = None
feat_path = MODEL_DIR / "features.json"
if feat_path.exists():
    try:
        FEATURES = json.load(open(feat_path)).get("features")
    except Exception:
        FEATURES = None

class PredictRequest(BaseModel):
    rows: List[Dict[str, Any]]
    threshold: Optional[float] = None  # optional override

class PredictResponse(BaseModel):
    probs: List[float]
    preds: List[int]
    threshold_used: float
    meta: Dict[str, Any]

def _to_dataframe(rows: List[Dict[str, Any]]) -> pd.DataFrame:
    df = pd.DataFrame(rows)
    # recode physiologically invalid values -> NaN (consistent with training)
    if "trestbps" in df: df.loc[df["trestbps"] <= 0, "trestbps"] = np.nan
    if "chol"     in df: df.loc[df["chol"]     <= 0, "chol"]     = np.nan
    if "oldpeak"  in df: df.loc[df["oldpeak"]   < 0, "oldpeak"]  = np.nan
    # align to training features if available
    if FEATURES:
        for c in FEATURES:
            if c not in df.columns:
                df[c] = np.nan
        df = df[FEATURES]
    return df

@app.on_event("startup")
def _load_model():
    global PIPE, CFG, THRESH
    PIPE = joblib.load(MODEL_DIR / "calibrated_pipeline.joblib")
    cfg_path = MODEL_DIR / "serve_config.json"
    if cfg_path.exists():
        CFG = json.load(open(cfg_path))
        THRESH = float(CFG.get("threshold", 0.5))

@app.get("/health")
def health():
    return {"status": "ok", "model": CFG.get("model"), "variant": CFG.get("variant")}

@app.post("/predict_proba", response_model=PredictResponse)
def predict_proba(req: PredictRequest):
    thr = float(req.threshold) if req.threshold is not None else THRESH
    X = _to_dataframe(req.rows)
    p = PIPE.predict_proba(X)[:, 1].tolist()
    preds = [int(pi >= thr) for pi in p]
    return PredictResponse(
        probs=p, preds=preds, threshold_used=thr,
        meta={"variant": CFG.get("variant"), "model": CFG.get("model")}
    )

@app.post("/predict", response_model=PredictResponse)
def predict(req: PredictRequest):
    return predict_proba(req)

if __name__ == "__main__":
    uvicorn.run("app:app", host="0.0.0.0", port=8080)
```
